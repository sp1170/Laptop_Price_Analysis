## Visualization

```python
df.head()
```

### Top 10 Laptop Brand by its Price

```python
plt.figure(figsize=(20,8))
df.groupby('Brand')["Final Price"].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='g')
plt.title("Top 10 Laptop Brand by its Price",size=30)
plt.ylabel("Laptop Price")
plt.show()
```

### Here , we can see that MSI brand is the most expensive Laptop brand.

### Top 10 Laptops by its Price

```python
plt.figure(figsize=(20,8))
df.groupby('Laptop')["Final Price"].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='gold')
plt.title("Top 10 Laptops by its Price",size=30)
plt.ylabel("Laptop Price")
plt.show()
```

### Here we can see , the most expensive and costly laptop is 'MSI Raider GE78HX 13VH-065ES Intel Core i9-13950HX/32GB/2TB SSD/RTX 4080/....'.

### Top 10 most expensive laptop model of MSI Brand 

```python
plt.figure(figsize=(20,8))
df[df['Brand']=='MSI'].groupby('Model')['Final Price'].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='g')
plt.title("Top 10 most expensive laptop model of MSI brand",size=30)
plt.ylabel("Laptop Price")
plt.show()
```

### Here we can see that , these are the MSI model in which Creator is  the most expensive and costly model of MSI.

### Top 10 Laptop by its RAM size

```python
plt.figure(figsize=(20,8))
df.groupby('Laptop')["RAM"].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='brown')
plt.title("Top 10 Laptop by its RAM size",size=30)
plt.ylabel("Laptop RAM")
plt.show()
```

###	MSI Titan GT77 HX 13VI-006ES Intel Core i9-139.. is the Laptop with highest RAM size i.e., 128GB

```python
plt.figure(figsize=(20,8))
df.groupby('Laptop')["Screen"].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='grey')
plt.title("Top 10 Laptops by its Screen size",size=30)
plt.ylabel("Laptop screen size")
plt.show()
```

### Here we can see that ASUS brand laptops have big screen size i.e., 17.5

```python
plt.figure(figsize=(20,8))
df.groupby('Laptop')["Storage"].sum().sort_values(ascending=False).head(10).plot(kind='bar',color='g')
plt.title("Top 10 Laptops by its Storage size",size=30)
plt.ylabel("Laptop Storage size in GB")
plt.show()
```

### ASUS ROG Zephyrus Duo 16 GX650RX-LO162W AMD Ryzen 9 6900HX/64GB/4TB SSD/RTX 3080 Ti/... has highest storage i.e. 4TB
